## iloc vs loc

#### iloc

Purely integer-location based indexing for selection by position.

`.iloc[]` is primarily integer position based (from 0 to length-1 of the axis), but may also be used with a boolean array.

Allowed inputs are:

* An integer, e.g. 5.

* A list or array of integers, e.g. [4, 3, 0].

* A slice object with ints, e.g. 1:7.

* A boolean array.

```python
df.iloc[:,:]
```

```python
df.iloc[0:3,:] # list[0:3] 0 is included and 3 is excluded 0,1,2 dislplay first 3 items head(3)
```

```python
df.iloc[0:3,]
```

```python
df.iloc[:,0:3]
```

```python
df.iloc[:,:]
```

### Handling Missing Values 

```python
data = {'Name':['Anil','Amir','Anali','Lasya','Ravi','Xi','Alapana',np.nan],
        'State':['TS','MH','UP','ML','TN','SK',np.nan,np.nan],
        'Gender':["M","M","F","F","M","M",np.nan,np.nan],      
        'Score':[63,48,56,75,np.nan,77,np.nan,np.nan]}
 
data = pd.DataFrame(data)
data
```

Drop all rows that have any NaN (missing) values

```python
data.dropna()
```

Drop only if entire row has NaN values

```python
data.dropna(how='all')
```

```python
data.dropna(thresh=2) # Drop a row if there are more than two NaN 
```

```python
data.dropna(subset=['Gender'])
```

Replace missing value with zeros

```python
data1 = data.fillna(0)
```

```python
data1
```

Replace missing value with Mean of the column

```python
data["Score"].fillna(data["Score"].mean(), inplace=True)
```

```python
data
```
